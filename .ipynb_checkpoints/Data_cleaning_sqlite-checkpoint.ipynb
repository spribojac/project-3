```python
import sqlite3
import pandas as pd
# Import dependencies.
import matplotlib
from matplotlib import style
style.use('fivethirtyeight')
import matplotlib.pyplot as plt
import chardet

# Python SQL toolkit and Object Relational Mapper
import sqlalchemy
from sqlalchemy.ext.automap import automap_base
from sqlalchemy.orm import Session
from sqlalchemy import create_engine, text, inspect, func
```

```python
csv_file_path = 'spotify_songs.csv'

# Read CSV file with Pandas
spotify_songs = pd.read_csv(csv_file_path, encoding='UTF-8')
```

```python
spotify_songs
```

```python
# Create SQLite connection
conn = sqlite3.connect('spotify_songs.sqlite')

# Write DataFrame to SQLite database
spotify_songs.to_sql("spotify_songs", conn, if_exists="replace", index=False)

# Close the connection
conn.close()
```

```python
engine = create_engine("sqlite:///spotify_songs.sqlite", echo=False)
# echo=False simply prevents sql out the sql statements being printed along with other data
```

```python
engine.execute(text('SELECT * FROM spotify_songs')).fetchall()
```
